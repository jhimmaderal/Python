```python
import pandas as pd
```

```python
df = pd.read_excel(r"Customer Call List.xlsx")
```

```python
# remove duplicates
dd = df.drop_duplicates()
```

```python
#remove columns
df = dd.drop(columns="Not_Useful_Column")
df
```

```python
#select column
# ----- Transformation -----
#lstrip - left side
#rstrip - right side
#strip - both side (remove delimeter in 1 line (.../,!))
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip(".../_")
df
```

```python
#eplace special characters to ''
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
# ----------------------#
# convert string
df["Phone_Number"] =df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df
```

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]= df["Phone_Number"].str.replace('Na--','')
df["Phone_Number"]
```

```python
df['Address']
df
#https://www.youtube.com/watch?v=bDhvCp3_lYw
```

```python
df[["Street_Address", "State", "Zip_Code"]]=df['Address'].str.split(',',2, expand = True)
df[["Street_Address", "State", "Zip_Code"]]
df
```

```python
#insert column
df.insert(3,"StreetAddress",df[["Street_Address"]])
```

```python
#remove columns
df = dd.drop(columns="Address")
df
```

```python
df["Paying Customer"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df["Paying Customer"]
```

```python
df["Do_Not_Contact"]= df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df["Do_Not_Contact"]
df
```

```python
df = df.replace('N/a',"")
df
```

```python
#fill NaN value
df.fillna("",inplace = True)
df
```

```python
df = df.replace('',"N")
df
```

```python
#1st Index
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x, inplace=True)
df
```

```python
# make new index
df.reset_index(drop=True)
```
